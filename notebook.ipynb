```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

Customers_df = pd.read_csv("customers_dataset.csv")
Geolocation_df = pd.read_csv("geolocation_dataset.csv")
Order_items_df = pd.read_csv("order_items_dataset.csv")
Order_payments_df = pd.read_csv("order_payments_dataset.csv")
Order_reviews_df = pd.read_csv("order_reviews_dataset.csv")
Orders_df = pd.read_csv("orders_dataset.csv")
Products_df = pd.read_csv("products_dataset.csv")
Sellers_df = pd.read_csv("sellers_dataset.csv")
product_category_name_translation_df = pd.read_csv("product_category_name_translation.csv")

```

```python
# Customers
Customers_df.info()
Customers_df.isna().sum()

print("Duplikasi = ", Customers_df.duplicated().sum())

Customers_df.describe()
Customers_df.describe(include='all')
```

```python
# Geolocation
Geolocation_df.info()
Geolocation_df.isna().sum()

Geolocation_df.describe(include='all')

# menghilangkan duplikasi
Geolocation_df.drop_duplicates(inplace=True)
print("Duplikasi = ", Geolocation_df.duplicated().sum())

Geolocation_df[Geolocation_df.geolocation_state.isna()]
Geolocation_df.geolocation_state.value_counts()
```

```python
# Order item
Order_items_df.info()
Order_items_df.isna().sum()

print("Duplikasi = ", Order_items_df.duplicated().sum())

Order_items_df.describe(include='all')
```

```python
# order payment
Order_payments_df.info()
Order_payments_df.isna().sum()

print("Duplikasi = ", Order_payments_df.duplicated().sum())

Order_payments_df.describe(include='all')
```

```python
# order review
Order_reviews_df.info()

Order_reviews_df.isna().sum()
Order_reviews_df[Order_reviews_df.review_comment_title.isna()]

# Mengisi missing value dengan teks placeholder
Order_reviews_df['review_comment_title'].fillna('No review', inplace=True)
Order_reviews_df['review_comment_message'].fillna('No comment', inplace=True)

print("Duplikasi = ", Order_reviews_df.duplicated().sum())

Order_reviews_df.describe(include='all')
```

```python
# orders
Orders_df.info()
Orders_df.isna().sum()

print("Duplikasi = ", Orders_df.duplicated().sum())

Orders_df.describe(include='all')

Orders_df[Orders_df.order_approved_at.isna()]
Orders_df.order_approved_at.value_counts()

# Mengubah kolom menjadi tipe datetime
datetime_columns = ["order_purchase_timestamp", "order_approved_at", "order_delivered_carrier_date", "order_delivered_customer_date", "order_estimated_delivery_date"]
for column in datetime_columns:
    Orders_df[column] = pd.to_datetime(Orders_df[column])

Orders_df.order_approved_at.fillna(Orders_df['order_approved_at'].median(), inplace=True)
Orders_df.order_delivered_carrier_date.fillna(Orders_df['order_delivered_carrier_date'].median(), inplace=True)
Orders_df.order_delivered_customer_date.fillna(Orders_df['order_delivered_customer_date'].median(), inplace=True)
```

```python
# product
Products_df.info()
Products_df.isna().sum()

print("Duplikasi = ", Products_df.duplicated().sum())

Products_df.describe(include='all')

Products_df.product_name_lenght.value_counts()
Products_df[Products_df.product_category_name.isna()]

# Menghapus baris dengan missing value pada kolom yang tersisa
Products_df.dropna(subset=['product_name_lenght', 'product_description_lenght', 'product_photos_qty', 'product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm'], inplace=True)
```

```python
# Product_category

product_category_name_translation_df.info()
product_category_name_translation_df.isna().sum()

print("Duplikasi = ", product_category_name_translation_df.duplicated().sum())
```

```python
# sellers
Sellers_df.info()
Sellers_df.isna().sum()

Sellers_df.describe(include='all')

print("Duplikasi = ", Sellers_df.duplicated().sum())
```

```python
# Analisis Distribusi Pelanggan Berdasarkan Kota dan Negara Bagian
# Menggabungkan Data Customers_df dengan Geolocation_df
# Menggabungkan data customers dengan geolocation berdasarkan zip_code_prefix
merged_customers = Customers_df.merge(Geolocation_df[['geolocation_zip_code_prefix', 'geolocation_city', 'geolocation_state']],
                                      left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix', how='left')

# Mengecek hasil gabungan
print(merged_customers.head())
```

```python
# Analisis Distribusi Berdasarkan Kota dan Negara Bagian
# Menghitung jumlah pelanggan per kota
city_distribution = merged_customers['geolocation_city'].value_counts().reset_index()
city_distribution.columns = ['city', 'customer_count']

# Menghitung jumlah pelanggan per negara bagian
state_distribution = merged_customers['geolocation_state'].value_counts().reset_index()
state_distribution.columns = ['state', 'customer_count']

# Menampilkan distribusi
print(city_distribution.head())
print(state_distribution.head())

# Visualisasi distribusi pelanggan berdasarkan kota (10 kota teratas)
plt.figure(figsize=(12, 6))
sns.barplot(x='customer_count', y='city', data=city_distribution.head(10))
plt.title('Top 10 Cities by Customer Count')
plt.xlabel('Customer Count')
plt.ylabel('City')
plt.show()

# Visualisasi distribusi pelanggan berdasarkan negara bagian
plt.figure(figsize=(12, 6))
sns.barplot(x='customer_count', y='state', data=state_distribution)
plt.title('Customer Distribution by State')
plt.xlabel('Customer Count')
plt.ylabel('State')
plt.show()
```

```python
# Analisis Penjualan Berdasarkan Kategori Produk yang Paling Laris dan yang Kurang Populer
# Menggabungkan data order_items dengan orders dan products
merged_order_items = Order_items_df.merge(Orders_df[['order_id', 'order_status']],
                                          on='order_id', how='left')

merged_order_items = merged_order_items.merge(Products_df[['product_id', 'product_category_name']],
                                              on='product_id', how='left')

merged_order_items = merged_order_items.merge(product_category_name_translation_df,
                                              on='product_category_name', how='left')

# Menampilkan hasil gabungan
print(merged_order_items.head())
```

```python
# Analisis Penjualan Berdasarkan Kategori Produk
# Filter hanya pesanan yang statusnya delivered
delivered_orders = merged_order_items[merged_order_items['order_status'] == 'delivered']

# Menghitung jumlah penjualan per kategori
category_sales = delivered_orders['product_category_name_english'].value_counts().reset_index()
category_sales.columns = ['product_category', 'sales_count']

# Menampilkan 10 kategori produk terlaris
print(category_sales.head(10))

# Menampilkan 10 kategori produk dengan penjualan terendah
print(category_sales.tail(10))

# Visualisasi
# Kategori produk terlaris
plt.figure(figsize=(12, 6))
sns.barplot(x='sales_count', y='product_category', data=category_sales.head(10))
plt.title('Top 10 Best-Selling Product Categories')
plt.xlabel('Sales Count')
plt.ylabel('Product Category')
plt.show()

# Kategori produk dengan penjualan terendah
plt.figure(figsize=(12, 6))
sns.barplot(x='sales_count', y='product_category', data=category_sales.tail(10))
plt.title('Top 10 Least-Selling Product Categories')
plt.xlabel('Sales Count')
plt.ylabel('Product Category')
plt.show()
```
